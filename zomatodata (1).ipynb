## Data Cleaning

The dataset is cleaned by removing duplicate records and preparing the data for analysis. Missing values and duplicate entries are checked to improve data quality.

```python
df.drop_duplicates(inplace=True)
```

## Visualization 1: Online Order Availability

This visualization shows the number of restaurants that provide online ordering services and those that do not.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="online_order")
plt.title("Online Order Availability")
plt.show()
```

## Visualization 2: Table Booking Availability

This graph illustrates how many restaurants offer table booking facilities.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="book_table")
plt.title("Table Booking Availability")
plt.show()
```

## Visualization 3: Restaurant Categories

This chart displays the different types of restaurants available in the dataset.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="listed_in(type)")
plt.xticks(rotation=45)
plt.title("Restaurant Types")
plt.show()
```

## Visualization 4: Votes vs Rating

This scatter plot shows the relationship between the number of customer votes and restaurant ratings. It helps identify whether restaurants with more customer votes tend to receive higher ratings.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="votes", y="rate")
plt.title("Votes vs Rating")
plt.show()
```

## Visualization 5: Distribution of Votes

This histogram represents the distribution of customer votes received by restaurants.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["votes"], bins=20)
plt.title("Distribution of Votes")
plt.show()
```
